```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
import sklearn
```

```python
df = pd.read_csv("irisyaya.csv",sep = ";")
```

```python
df
```

```python
df[['P_l','P_W','S_L','S_W']]
```

```python
df[['Output1','Output2','Output3']]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df['P_l']=(df['P_l']- min(df['P_l']))/(max(df['P_l'])-min(df['P_l']))
```

```python
df
```

```python
for i in ['P_W','S_L','S_W']:
   df[i] = (df[i]- min(df[i]))/(max(df[i])-min(df[i]))
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df[['P_l','P_W','S_L','S_W']], df[['Output1','Output2','Output3']], random_state=42, test_size = 0.3)
```

```python
X_train
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(10,),solver='adam',learning_rate_init= 0.1, max_iter=500, activation = "logistic", alpha = 0.001, momentum = 0.1) 
```

```python
mlp.fit(X_train, y_train)
```

```python
mlp.score(X_test,y_test)
```
